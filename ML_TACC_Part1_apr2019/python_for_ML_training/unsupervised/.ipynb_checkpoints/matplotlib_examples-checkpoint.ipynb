## Random Walk 1-D

```python
%matplotlib inline
import random
import numpy as np
import matplotlib.pyplot as plt

# Start at Zero
position = 0

# List to capture each step
walk = list()

# Number of steps
steps = 1000

# Run 1000 steps (iterations)
for i in np.arange(steps):
    # Coin toss
    if random.randint(0, 1):
        step = 1 
    else:
        step = -1
    # Add step
    position += step
    # Append to walk list
    walk.append(position)
    
# Setup X and Y values
x = np.arange(steps)
y = np.array(walk)

# Plot the random walk
plt.plot(x,y)
```

## Random Walk 3-D

```python
%matplotlib inline
import random
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


# Start at Zero
position_x = 0
position_y = 0
position_z = 0

# List to capture each step
walk_x = list()
walk_y = list()
walk_z = list()

# Number of steps
steps = 1000

# Run 1000 steps (iterations)
for i in np.arange(steps):
    # For x,y,z
    for i in (1,2,3):
        # Coin toss
        if random.randint(0, 1):
            step = 1 
        else:
            step = -1
        # Add step
        if(i==1):
            position_x += step
            # Append to walk list
            walk_x.append(position_x)
        elif(i==2):
            position_y += step
            # Append to walk list
            walk_y.append(position_y)
        else:
            position_z += step
            # Append to walk list
            walk_z.append(position_z)
                    
# Setup X and Y values
#x = np.arange(steps)
x = np.array(walk_x)
y = np.array(walk_y)
z = np.array(walk_z)


# Plot the random walks individually
plt.plot(np.arange(steps),x)
plt.plot(np.arange(steps),y)
plt.plot(np.arange(steps),z)

# Plot 3d random walk
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot(x, y, z, label='random walk 3d')
ax.legend()
```
