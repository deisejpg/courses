```python
#Import Numpy
import numpy as np
#Import SciPy FFT
from scipy.fftpack import fft
#Import Matplotlib
import matplotlib.pyplot as plt
```

```python
# Number of sample points
N = 600
```

```python
#sampling interval
T = 1.0 / 800.0
```

```python
# Xaxis points
x = np.linspace(0.0, N*T, N)
```

```python
#Yaxis points (Adding 2 sinusoidal signals 50Hz and 80Hz)
y = np.sin(50.0 * 2.0 * np.pi * x) + 0.5 * np.sin(80.0*2.0*np.pi*x)
```

```python
# plot original signal
plt.plot(x, y)
plt.grid()
plt.show()
```

```python
# Y and X frequency domain
yf = fft(y)
xf = np.linspace(0.0, 1.0/(2.0*T), N/2)
```

```python
# plot frequecies
plt.plot(xf, 2.0/N * np.abs(yf[0:N/2]))
plt.grid()
plt.show()
```
